#### 요구사항 3. 다음과 같이 step에 주어진 증분만큼 count를 증가시키는 incr() 메소드를 Counter 클래스 내에 구현하시오.

#### 요구사항 5. 다음과 같은 두 개의 산술 연산 (+, -)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

> \>>> c = c + 5 

> \>>> d = d - 5 

> \>>> print(c) 

> [Count (step: 1)] 17 

> \>>> print(d) 

>[Count (step: 2)] 9 

#### 요구사항 6. 다음과 같은 관계연산 (>, <, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

> \>>> print(c) > 10 

>True 

> \>>> print(d) > 10 

>False 

> \>>> print(c) < 10 

>False 

> \>>> print(d) < 10 

>True 

> \>>> print(c) == 17 

>True 

> \>>> print(d) != 9 

>False 

```python
class Counter:
    def __init__(self, count, step=1):
        self.count = count
        self.step = step
    def __str__(self):
        return str(self.count)
    def incr(self):
        self.count = self.count + self.step
        return self.count
    def __call__(self):
        self.incr()
        return str(self.count)
    def __add__(self, B):
        return self.count + B
    def __sub__(self, B):
        return self.count - B
    def __gt__(self, y):
        return self.count > y
    def __lt__(self, y):
        return self.count < y
    def __eq__(self, y):
        return self.count == y
    def __ne__(self, y):
        return self.count != y
    
c = Counter(10)
d = Counter(10,2)
c.incr()
d.incr()
c()
d()
c = c + 5
d = d - 5
print(c>10)
print(d>10)
print(c<10)
print(d<10)
print(c==17)
print(d!=9)
```

일단 생성자로 count값과 step값을 인자로 받을 수 있게 하고(step의 default 값: 1), 출력을 해주는 \__str\__() 메소드, step에 주어진 증분만큼 count를 증가시키는 incr() 메소드, 인스턴스 객체를 직접 호출할 수 있게 해주는 \__call\__() 메소드, 덧셈뺄셈을 가능하게 하는 \__add\__(), \__sub\__() 메소드, 관계연산을 수행할 수 있는 \__gt\__(), \__lt\__(), \__eq\__(), \__ne\__() 메소드를 구현했다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

먼저 s = MySet([1,2,2,3]) 에서 클래스의 인스턴스를 생성하면서 생성자가 호출된다. 인자로 안에 있는 리스트를 받고 for~in 문으로 자기 자신의 리스트에 append 메소드로 인자로 받은 리스트 안에 있는 것을 하나씩 붙인다. 그리고는 MySet.eliminate_duplicate(self) 로 인자로 자기 자신인 리스트를 주고 정적 메소드인 eliminate_duplicate 를 호출한다. 그러면 for~in 문 안에서 if e not in s: 로 인자로 받은 리스트에서 중복되는 것을 제외해서 새로운 리스트를 만든 후, 이를 인자로 받은 리스트에 다시 집어넣는다. 마지막으로 \__str\__ 메소드를 불러 출력하는 데 있어서 형식을 갖춰 출력하게 한다. 즉 MySet 클래스를 사용해서 리스트를 집합(중복되는 것이 없음)으로 만들 수 있다.

### [5번 문제]
4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오

- 정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.
- 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.

> \>>> u = s | t 

> \>>> print(u) 

> MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9} 

- 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.

> \>>> u = s & t 

> \>>> print(u) 

> MySet: {2, 3} 

- 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.

> \>>> s = MySet([1, 2, 3]) 

> \>>> t = MySet([3, 4, 5]) 

> \>>> u = s - t 

> \>>> print(u) 

> MySet: {1, 2} 

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result
    
    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
    # 합집합 연산
    # 한 집합에다가 다른 집합의 원소들을 추가하고 중복되는 것들을 제거
    def __or__(self,B):
        orList = MySet([])
        for x in self:
            orList.append(x)
        for x in B:
            orList.append(x)
        MySet.eliminate_duplicate(orList)
        return orList
    # 교집합 연산
    # 두 집합에 동일하게 들어있는 원소만을 반환
    def __and__(self,B):
        andList = MySet([])
        for y in B:
            if y in self:
                  andList.append(y)
        return andList 
    # 차집합 연산
    # 다른 집합에 동일한 원소가 존재하면 이를 제거
    def __sub__(self, B):
        subList = MySet([])
        for x in self:
            subList.append(x)
        for y in B:
            if y in subList:
                subList.remove(y)
        return subList
    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3, 6])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9,9])
    print(t)
    print()
    w = s|t
    print(w)
    u = s&t
    print(u)
    m = s-t
    print(m)
```

합집합은 단순히 다른 집합에 있는 것들을 추가해서 중복을 없애면 되고, 교집합은 동일하게 들어있는 것들을 반환, 차집합은 한 집합에 있는게 다른 집합에도 있다면 이를 제거해주면 된다.

```python
class OrderedWebWordsFrequency(WebWordsFrequency):
    def __init__(self,*args):
        WebWordsFrequency.__init__(self,*args)
    def getWordsFrequency(self, reverse = False):
        M = []
        L = []
        m = {}
        p = []
        dM = {}
        Rev = False
        if reverse == False:
            Rev = True
        
        for x in self.urlList:
            M.append(htmlSiteDict(x))
        for x in M:
            for key in x.keys():
                for y in range(x[key]):
                    L.append(key)
        for k in L:
            m[k] = L.count(k)
        dM = sorted(m.items(),key = operator.itemgetter(1),reverse = Rev)
        return dM
    def __iter__(self):
        self.index = 0
        self.p = self.getWordsFrequency()
        return self
    def __next__(self):
        if self.index >= len(self.p):
            raise StopIteration
        n = self.p[self.index]
        self.index += 1
        return n
w4 = OrderedWebWordsFrequency('http://www.times.com', 'https://github.com')
for i in w4:
    print(i)
```
